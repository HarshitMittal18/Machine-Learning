#### K-Nearest Neighbours Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

##### Data Preparation

```python
# Generating Data from Distributions
mean1 = np.array([3,3])
cov1 = np.array([[1.2,-0.5],[-0.5,1]])

mean2 = np.array([4,5])
cov2 = np.array([[1,0.5],[0.5,1.2]])

dist1 = np.random.multivariate_normal(mean1,cov1,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(X1.shape)
print(X2.shape)
```

```python
# Plot Distributions
plt.scatter(dist1[:,0],dist1[:,1], label = 'class 0')
plt.scatter(dist2[:,0],dist2[:,1], label = 'class 1')
# Query Point
query_point = np.array([3,4.0])
plt.scatter(query_point[0],query_point[1],color = 'red', marker = '*', s = 300)
plt.legend()
plt.show()
```

```python
# Creating a Single Dataset
rows = dist1.shape[0] + dist2.shape[0]
cols = dist1.shape[1] + 1

data = np.zeros((rows,cols))

data[:dist1.shape[0],:2] = dist1
data[:dist1.shape[0],-1] = 0
data[dist2.shape[0]:,:2] = dist2
data[dist2.shape[0]:,-1] = 1

# Shuffling the Dataset
np.random.shuffle(data)

print(data.shape)
```

```python
# Creating Training and Testing Data
split = int(0.8*data.shape[0])
# Training Data
Xtrain = data[:split,:-1]
Ytrain = data[:split,-1]
# Testing Data
Xtest = data[split:,:-1]
Ytest = data[split:,-1]
```

```python
print(Xtrain.shape,Ytrain.shape,Xtest.shape,Ytest.shape)
```

##### Algorithm

```python
class KNN:
    
    def __init__(self,K = 5):
        self.K = K
        
    def fit(self,Xtrain,Ytrain):
        self.Xtrain = Xtrain
        self.Ytrain = Ytrain
    
    def euclidean_distance(self,x1,x2):
          return np.sqrt(sum((x1-x2)**2))
    
    def predict(self, query_point):
        # Calculate Distance
        distances = []
        for data_point,label in zip(Xtrain,Ytrain):
            distance = self.euclidean_distance(data_point,query_point)
            distances.append((distance,label))

        # Sorting 
        distances = sorted(distances,key=lambda x:x[0])
                             
        # Finding K Nearest Neighbours
        neighbours = np.array(distances[:self.K])
                             
        # Majority Vote. In case of regression use weighted mean
        labels,counts = np.unique(neighbours[:,1],return_counts=True)
        pred = labels[np.argmax(counts)]
                             
        return pred
    
    def score(self, Xtest, Ytest):
        # Get Predictions for Test Data
        predictions = []
        for test_point in Xtest:
            prediction = self.predict(test_point)
            predictions.append(prediction)
            
        # Calculating Accuracy
        predictions = np.array(predictions)
        return sum(Ytest == predictions)/Ytest.shape[0]
```

```python
# Initializing Object
Obj = KNN()
```

```python
# Fitting Training Data to Algorithm
Obj.fit(Xtrain,Ytrain)
```

```python
# Getting Prediction for Testing Point
Obj.predict(query_point)
```

```python
# Getting Model Accuracy
Obj.score(Xtest,Ytest)
```

#### Change of Test Accuracy with 1/K

```python
Error = []
for K in range(1,101):
    # Create Object
    Obj = KNN(K)
    # Fit Training Data
    Obj.fit(Xtrain,Ytrain)
    # Generate Accuracy
    accuracy = Obj.score(Xtest,Ytest)
    print("Error for K = {}: {:.2f}".format(K,1-accuracy))
    Error.append(1-accuracy)

# Plot Graph     
plt.plot(1/np.arange(1,101),Error)
plt.show()
```
