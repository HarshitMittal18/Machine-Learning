```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
X = np.arange(10)
Y = (X-5)**2
print(X)
print(Y)
```

```python
# Visualization
plt.style.use('dark_background')
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('F(X)')
plt.show()
```

<small>Goal: Given a function F(X), we want to find the value of X at which F(X) is minimum



```python
# Initial Value 
x = 0 
# Learning Rate
lr = 0.1

Error = []
# We will take 50 steps in the downward direction 
plt.plot(X,Y)
for i in range(50):
    gradient = 2*(x-5)
    x = x - (lr*gradient)
    y = (x-5)**2
    Error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
# Plot the values for error
plt.plot(Error)
plt.show()
```
