```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Data Preparation

```python
# Sample out data from normal distribution
mean1 = [1,1]
cov1 = [[0.1,0],[0,0.1]]
dist1 = np.random.multivariate_normal(mean1,cov1,100)

mean2 = [1.5,4]
cov2 = [[0.1,0],[0,0.1]]
dist2 = np.random.multivariate_normal(mean2,cov2,100)

mean3 = [0.5,2.5]
cov3 = [[0.1,0],[0,0.1]]
dist3 = np.random.multivariate_normal(mean3,cov3,100)

mean4 = [2,2]
cov4 = [[0.1,0],[0,0.1]]
dist4 = np.random.multivariate_normal(mean4,cov4,100)
```

```python
# Combine Data
dataset = np.zeros((dist1.shape[0]*4, dist1.shape[1]))
dataset[0:dist1.shape[0],:] = dist1
dataset[dist2.shape[0]:dist3.shape[0]*2,:] = dist2
dataset[dist3.shape[0]*2:dist4.shape[0]*3,:] = dist3
dataset[dist4.shape[0]*3:,:] = dist4
print(dataset.shape)
```

#### Algorithm

```python
# Cluster Information
class Cluster:
    def __init__(self,center,color):
        self.center = center
        self.color = color
        self.points = list()
    
    # Returns a user friendly string representation
    def __str__(self):
        return "Cluster <{}>".format(self.color)
    
    # Return a developer friendly string representation
    def __repr__(self):
        return "Cluster({},'{}')".format(self.center,self.color)
    
    def update_center(self):
        if not self.points:
            return self.center
        
        new_center = np.array(self.points).mean(axis = 0)
        self.center = new_center
        return new_center
```

```python
class KMeansClustering:
    def __init__(self,K,color,dataset):
        
        self.K = K 
        self.color = color
        self.dataset = dataset
        self.inertia = list()
        
        # Initializing Cluster Centers
        clusters = []
        random_centers = np.random.uniform(low = dataset.min(), high = dataset.max(), size = (K,dataset.shape[1]))
        
        for center,color in zip(random_centers,self.color):
            clusters.append(Cluster(center,color))
        
        self.clusters = clusters
        self.centers = random_centers
        
        # Visualize Initial State
        plt.figure()
        plt.style.use('dark_background')
        plt.scatter(self.dataset[:,0],self.dataset[:,1])
        plt.scatter(self.centers[:,0],self.centers[:,1], color = 'white', marker = '*', s = 200)
        plt.title("Visualize Initial State")
        plt.show()

        
    def run(self, iterations = 1):
        
        for _ in range(iterations):
            self.assign_closest_cluster()
            self.update_cluster_center()
            self.inertia.append(self.calculate_inertia())
            
    def assign_closest_cluster(self):
        
        for cluster in self.clusters:
            cluster.points = list()
        
        for point in self.dataset:
            
            distances = []
            
            for cluster in self.clusters:
                distance = KMeansClustering.euclidean(point,cluster.center)
                distances.append((distance,cluster))
            
            distances = sorted(distances, key = lambda x:x[0])
            closest_center = distances[0][1]
            closest_center.points.append(point)
            
    def update_cluster_center(self):
        new_centers = []
        for cluster in self.clusters:
            new_centers.append(cluster.update_center())
        
        self.centers = new_centers
        
    def calculate_inertia(self):
        inertia = 0
        for cluster in self.clusters:
            for point in cluster.points:
                inertia += KMeansClustering.euclidean(cluster.center,point)
        return inertia
    
    def plot_iteratively(self,iterations,inp=True):
        
        for iteration in range(iterations):
            self.run()
            plt.figure()
            plt.style.use('dark_background')
            for cluster in self.clusters:
                points = np.array(cluster.points)
                plt.scatter(cluster.center[0], cluster.center[1], color='white', marker='*', s=200)
                if not cluster.points:
                    continue
                plt.scatter(points[:, 0], points[:, 1], color=cluster.color, alpha = 0.8)
            
            plt.show()
            if inp:
                input()
                
    def plot_inertia(self):
        plt.figure()
        plt.title("Inertia VS Iterations")
        plt.plot(self.inertia)
        plt.show()
                        
    @staticmethod
    def euclidean(p1, p2):
        return np.sqrt(np.sum((p1 - p2)**2))
```

```python
# Creating KMeansClustering Object
K = 4
colors = ['red', 'green', 'blue', 'yellow', 'pink']
Kmeans = KMeansClustering(K,colors, dataset)
```

```python
# Plotting clusters iteratively
Kmeans.plot_iteratively(5,inp=False)
```

```python
# Plotting Inertia
Kmeans.plot_inertia()
```
