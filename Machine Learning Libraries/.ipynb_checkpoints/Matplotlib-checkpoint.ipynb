### Matplotlib

Matplotlib is a plotting library for the Python programming language

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Themes

<small>We can change the background theme of a Matplotlib Graph

```python
# To print the available themes
themes = plt.style.available
print(themes)
```

```python
# To set a particular theme
plt.style.use('dark_background')
```

#### Line Plot

```python
# Dummy Data
x = np.arange(0,10)
y1 = x**2
y2 = 2*x + 3
print(x,y1,y2,sep = '\n')
```

```python
# Line PLot
plt.plot(x,y1,color = 'blue', label = 'Apple', linestyle = 'dashed') # Label for Legend
plt.plot(x,y2,color = 'yellow', label = 'Banana', marker = 'o') # Label for Legend
# X-Axis Label
plt.xlabel('Time')
# Y-Axis Label
plt.ylabel('Price')
# Title 
plt.title("Prices of Fruits over time")
# Plot Legend as defined by the label
plt.legend()
# plot.show() has two usage:
#1 Necessary to show the plot in the pop up window 
#2 Marks the end of a plot 
plt.show() 
```

<small>Note:
In case we pass a single array to the plot function, it will take the array passed as ylabel and use the indices of that array as xlabels to generate a line plot

```python
plt.plot(y1, marker = 'o')
plt.show()
```

#### Scatter Plot

```python
# To adjust the size of the plot
plt.figure(figsize=(7,4))
# Scatter Plot
plt.scatter(x,y1,color = 'blue', label = 'Apple', marker = "*")
plt.scatter(x,y2,color = 'yellow', label = 'Banana', marker = 'o')
# X-Axis Label
plt.xlabel('Time')
# Y-Axis Label
plt.ylabel('Price')
plt.title("Prices of Fruits over time")
# Plot Legend as defined by the label
plt.legend()
plt.show() 
```

#### Bar Plot

```python
# Bar Plot
plt.bar(x = ['Apple','Orange','Banana'], height = [20,10,15]) # Current Year
plt.show()
```

```python
# Code to create side by side bar plot
# Choose the x-coordinates to place the bar and adjust the width to remove the overlap between bars
x_cordinate = np.array([1,2,3])*2
# Current Year. 
plt.bar(x = x_cordinate, height = [20,10,15],width = 0.5, label = "Current Year", tick_label = 
        ['Gold','Platinum','Silver'])
# Previous Year
plt.bar(x = (x_cordinate + 0.5), height = [15,10,10], width = 0.5, label = "Previous Year") 
plt.title('Metals Price Comparison')
plt.xlabel('Metals')
plt.ylabel('Price')
# Puts a limit on y-axis
plt.ylim(0,30)
# Puts a limit on x-axis
plt.xlim(1,8)
plt.legend()
plt.show()
```

#### Pie Chart

```python
subjects = 'Maths','English','History','Science'
# The sum of the weights is not required to be 100
weights = [20,20,30,30]
# Explode: Takes the radius as input. Helps us to differentiate a particular category 
# Autopct: Displays the labels in percentage
plt.pie(weights,labels=subjects, explode = (0,0,0,0.2), autopct = '%1.1f%%')
plt.show()
```

#### Histogram

```python
# Generating data with mean = 60 and std = 5 from a standard normal distribution
# Data from Standard Normal Distribution
Xsn = np.random.randn(100)
# Mean and Standard Deviation
u = 60
std = 5
# Empirical Formula
X1 = np.round(Xsn * std + u)
X2 = np.round(Xsn * std + (u-20))
print(X1)
print(X2)
```

```python
# Histogram
# Alpha: Manges the opacity of the plot
plt.hist(X1,alpha = 0.8, bins = 20)
plt.hist(X2,alpha = 0.8, bins = 20)
plt.xlabel("Marks Range")
plt.ylabel("Frequency Count")
plt.show()
```
