```python
import numpy as np 
import matplotlib.pyplot as plt
```

#### Univariate Normal Distribution

<small>Standard Normal Distribution with mean = 0 and std = 1

```python
# Data 
data = np.random.randn(1000)
# Histogram
plt.style.use('dark_background')
plt.hist(data,20)
plt.show()
```

<small>Normal Distribution with mean = 7 and std = 70

```python
X = data * 7 + 70
plt.hist(X,20)
plt.show()
```

```python
plt.scatter(X,np.zeros(X.shape))
plt.show()
```

#### Multivariate Normal Distribution

<small>Bivariate Normal Distribution

```python
# Data Generation
mean1 = np.array([0.0,0.0])
cov1 = np.array([[1,-0.5],[-0.5,1]])

mean2 = np.array([4,5])
cov2 = np.array([[1.2,0.5],[0.5,1.2]])

dist1 = np.random.multivariate_normal(mean1,cov1,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist1.shape)
print(dist2.shape)
```

```python
# Plot
plt.scatter(dist1[:,0],dist1[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
